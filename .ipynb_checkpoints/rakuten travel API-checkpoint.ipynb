## rakuten travel API

```python
import requests
```

```python
URL = 'https://app.rakuten.co.jp/services/api/Travel/KeywordHotelSearch/20170426'
APP_ID = '1076743242127031162'
```

```python
params = {
    'applicationId': APP_ID,
    'keyword': '沖縄'
}
```

```python
#APIをたたく（APIを実行）

res = requests.get(URL, params)
```

```python
#アクセス成否確認

res.status_code
```

```python
#取得情報

res.json()
```

```python
#取得情報の掘り下げ

hotels = res.json()['hotels']
hotels[0]
```

```python
#取得情報の掘り下げ

hotel = hotels[0]['hotel'][0]['hotelBasicInfo']
hotel
```

```python
#取得情報の掘り下げ

hotel['access']
```

```python
#取得情報の掘り下げ

hotel['hotelName']
```

```python
#取得情報の掘り下げ

address = hotel['address1'] + hotel['address2']
address
```
